```python
data = pd.read_csv('Solar Power Energy  by Country.csv')
solar = data.copy()
```

```python
#Eliminar filas innecesarias
solar = solar.fillna(0)
solar = solar.drop(['2019.1', '2021.1','Share of totalconsumption','Unnamed: 0'], axis=1)
solar = solar.drop(index=0, axis=0)
solar = solar.dropna()

#Renombrar columnas
solar.rename(columns={'Unnamed: 13': 'Total_final'}, inplace=True)
solar.rename(columns={'Unnamed: 3': 'Total_2016'}, inplace=True)
solar.rename(columns={'Unnamed: 5': 'Total_2017'}, inplace=True)
solar.rename(columns={'Unnamed: 7': 'Total_2018'}, inplace=True)
solar.rename(columns={'Unnamed: 9': 'Total_2019'}, inplace=True)
solar.rename(columns={'Unnamed: 11': 'Total_2020'}, inplace=True)

solar.rename(columns={'2016': 'Nuevo_2016'}, inplace=True)
solar.rename(columns={'2017': 'Nuevo_2017'}, inplace=True)
solar.rename(columns={'2018': 'Nuevo_2018'}, inplace=True)
solar.rename(columns={'2019': 'Nuevo_2019'}, inplace=True)
solar.rename(columns={'2020': 'Nuevo_2020'}, inplace=True)
solar.rename(columns={'2021': 'Nuevo_2021'}, inplace=True)


#Fijamos la primera columna como Ã­ndice y la ponemos en snake case y minÃºscula. Pasamos a Float
solar.set_index(solar.columns[0], inplace=True)
solar = solar.apply(pd.to_numeric, errors='coerce').astype(float)
```

## RepresentaciÃ³n de los mayores productores de EnergÃ­a solar

```python
solar15 = solar.head(20)

total_final_column = solar15['Total_2016']

# Crear el histograma con el Ã­ndice como eje x
plt.barh(solar15.index, total_final_column, color=['red', 'green', 'blue', 'yellow', 'orange', 'purple', 'pink', 'brown', 'cyan', 'magenta', 'lime', 'teal', 'lavender', 'maroon', 'navy', 'olive', 'silver', 'black', 'white', 'gray'])

# Etiquetas y tÃ­tulo
plt.xlabel('Total Final')
plt.ylabel('PaÃ­s')
plt.title('EnergÃ­a solar producida 2016')

# Rotar etiquetas del eje x para mayor legibilidad
plt.xticks(rotation=90)

# Mostrar el histograma
plt.tight_layout()
plt.show()
```
